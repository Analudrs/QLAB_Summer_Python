#### 1.1 Create a dictionary with two keys: even_numbers and odd_numbers. The first key should have all the even numbers in this range [0, 2000], and the second key must have all the odd numbers in this range [9000, 19000]. The values should be stored in a list. Hint: Use the np.arange, zip, and np.tolist() functions.

```python
import numpy as np 
```

```python
even_numbers = np.arange(0, 2000, 2).tolist()
even_numbers 
```

```python
odd_numbers = np.arange(9001, 19000, 2).tolist()
odd_numbers
```

```python
z = list(zip(even_numbers, odd_numbers))
print(z)
```
