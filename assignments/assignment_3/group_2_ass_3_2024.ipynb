```python
import pandas as pd

# Load the Excel file
xls = pd.ExcelFile('./_data/bogota_firms.xlsx')

# Get all sheet names
sheet_names = xls.sheet_names
print(f"All sheet names: {sheet_names}")

# Load a specific sheet to a DataFrame by its index (0 is the first sheet, 1 is the second sheet, etc.)
df = pd.read_excel(xls, sheet_name=1)

# Display the first few rows of the DataFrame to inspect it
print(df.head())







```

```python
# Show some descriptive statistics
print(df.describe())
```

```python


# Drop columns if they are in Dollars
df = df.loc[:, ~df.columns.str.contains('USD')]

# Define the columns you want to move to the end
cols_to_move = ['Company name Latin alphabet'] + ['Operating revenue (Turnover)\nth LCU 2023']

# Other columns
other_cols = df.columns.difference(cols_to_move)

# Move the columns to the end
df = pd.concat([df.drop(cols_to_move, axis=1), df[cols_to_move]], axis=1)

print(df.head())
```

```python
# Define a dictionary to hold the new names
new_names = {}

# Loop over the years you're interested in
for year in range(2011, 2024):
    old_name = f'Operating revenue (Turnover)\nth LCU {year}'
    new_name = f'Op. Revenue {year}'
    new_names[old_name] = new_name

# Add other column renames
new_names['Company name Latin alphabet'] = 'Firm Name'
new_names['Country ISO code'] = 'ISO Code'
new_names['City\nLatin Alphabet'] = 'City Name'
new_names['NACE Rev. 2, core code (4 digits)'] = 'NACE Rev.2 Code'
new_names['Consolidation code'] = 'Cnslidation Code'
new_names['Last avail. year Op.'] = 'Last year Op.'
# Rename the columns
df.rename(columns=new_names, inplace=True)

print(df.head())
```
