```python
import pandas as pd

# Especifica la ruta del archivo Excel
ruta_del_archivo_excel = '../../_data/bogota_firms.xlsx'

# Usa la función read_excel para cargar los datos en un DataFrame
data_frame = pd.read_excel(ruta_del_archivo_excel, sheet_name='Results')

# Ahora puedes trabajar con tu DataFrame
# Por ejemplo, puedes imprimir las primeras filas del DataFrame
data_frame.head(5)

# Limpiando base de datos
# Reemplaza los valores nulos (NaN) por 0 en todo el DataFrame
data_frame_sin_na = data_frame.replace( "n.a.", 0)
data_frame_sin_na.head(5)
data_frame_sin_na.describe(include="all").round(4)
```

# 2


```python
#las columnas con las que nos vamos a quedar serán el nombre de la empresa y también conservaremos los años. Para eso usamso el siguiente código:
cols_to_keep = ['Company name Latin alphabet']
for a in range(2011,2024):
    revenue_dolar="Operating revenue (Turnover)\nth LCU {}".format(a)
    cols_to_keep.append(revenue_dolar)
#guardamos estso datos como una nuevo data frame
newdf = data_frame_sin_na[cols_to_keep]
newdf
```

# 3

```python
# Para renombrar las columnas con nombres más cortos, definimos
short_names = {}

short_names['Company name Latin alphabet'] = 'firm_name'

# Usamos un loop para las columnas de los resultados operativos
for n in range(2011, 2024):
    long_name = f'Operating revenue (Turnover)\nth LCU {n}'
    short_name = f'oper_rev_{n}'
    short_names[long_name] = short_name

# Finalmente, usamos el siguiente código para cambiar los nombres
newdf.rename(columns=short_names, inplace=True)
newdf.head()
```
